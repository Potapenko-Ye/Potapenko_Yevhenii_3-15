```python
import matplotlib.pyplot as plt
import seaborn as sns
mtx = df.drop('DEATH_EVENT', axis=1).corr(numeric_only=True).abs()

# Побудова теплової карти кореляції
fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),  # маскуємо верхній трикутник
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
corr_with_EVENT = df.corr()['DEATH_EVENT'].abs().sort_values(ascending=False)
corr_with_EVENT
```

```python
df['DEATH_EVENT'].value_counts()
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
x = df.drop(columns=['DEATH_EVENT'])
y = df['DEATH_EVENT']
discrete_features = x.columns.isin([
    'anaemia', 'diabetes', 'high_blood_pressure',
'sex', 'smoking'])
mi_scores = mutual_info_classif(
    x,y,
    discrete_features=discrete_features,
    random_state=42
)
mi_scores = pd.Series(
    mi_scores,
    name = 'MI Scores',
    index=x.columns
).sort_values()
print(mi_scores.sample(5))
```

```python
nonzero_mi_scores = mi_scores[mi_scores > 0].sort_values(ascending=False)
top_mi_scores = nonzero_mi_scores.head(7)
plt.figure(figsize=(8,6))
plt.barh(top_mi_scores.index[::-1], top_mi_scores[::-1])
plt.title('Top Mutual Information Scores')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```

```python
from google.colab import files
import pandas as pd
import numpy as np
import os
uploaded = files.upload()
```

```python
df = pd.read_csv("taxi_trip_pricing.csv")

df
```

```python
df.info()
```

```python
df['Time_of_Day'].value_counts()
```

```python
df['Day_of_Week'].value_counts()
```

```python
df['Traffic_Conditions'].value_counts()
```

```python
df['Weather'].value_counts()
```

```python
df['Traffic_Conditions'] = df['Traffic_Conditions'].map({
    'Low': 0,
    'Medium': 1,
    'High': 2
})

df['Weather'] = df['Weather'].map({
    'Clear': 0,
    'Rain': 1,
    'Snow': 2
})

df['Day_of_Week'] = df['Day_of_Week'].map({
    'Weekday': 0,
    'Weekend': 1,
})

df['Time_of_Day'] = df['Time_of_Day'].map({
    'Afternoon': 0,
    'Morning': 1,
    'Evening': 2,
    'Night': 3
})
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
mtx = df.drop('Trip_Price', axis=1).corr(numeric_only=True).abs()

# Побудова теплової карти кореляції
fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),  # маскуємо верхній трикутник
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
corr_with_EVENT = df.corr()['Trip_Price'].abs().sort_values(ascending=False)
corr_with_EVENT
```

```python
df['Trip_Price'].value_counts()
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
x = df.drop(columns=['Trip_Price'])
y = df['Trip_Price']
discrete_features = x.columns.isin([
    'Traffic_Conditions', 'Weather', 'Day_of_Week',
'Time_of_Day'])
mi_scores = mutual_info_regression(
    x,y,
    discrete_features=discrete_features,
    random_state=42
)
mi_scores = pd.Series(
    mi_scores,
    name = 'MI Scores',
    index=x.columns
).sort_values()
print(mi_scores.sample(5))
```

```python
nonzero_mi_scores = mi_scores[mi_scores > 0].sort_values(ascending=False)

top_mi_scores = nonzero_mi_scores.head()

plt.figure(figsize=(8, 6))
plt.barh(top_mi_scores.index[::-1], top_mi_scores[::-1])
plt.title('Top Munual Information Scores')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```
